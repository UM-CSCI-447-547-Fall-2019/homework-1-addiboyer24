We're primarily interested in determining the relationship between relative frequency and magnitude.  Extract the magnitude variable from the pandas data frame using the 'mag' key.  

Next, you'll need to determine the number of earthquakes that occured in the data as a function of magnitude.  This is easily done by deciding on a set of bins, and then counting the number of items in each bin.  This is also known as a [histogram](https://docs.scipy.org/doc/numpy/reference/generated/numpy.histogram.html), and is easily computed using either numpy or matplotlib.  You'll want to use a fairly large number of bins, say 50 (equally spaced).  *Note that the histogram function provided by either of the above libraries returns total counts rather than relative frequencies: you'll need to compute relative frequencies by dividing the returned counts by the total number of earthquakes in the dataset.*

```python
import numpy as np
import matplotlib.pyplot as plt
#Magnitude
x = data['mag']

# Bin the data
num_bins = 50
h_bins = []
bin_size = ((max(x)-min(x)) / num_bins)
start = min(x)
for i in range(num_bins):
    h_bins.append(start)
    start+=bin_size

hist = np.histogram(x, bins=h_bins)
print(hist[0])
print(hist[1])
plt.hist(x, bins=h_bins)
plt.show()


```

**(20 pts) Make a plot with bin centroids on the $x$-axis and relative frequency on the $y$-axis.  Based on your results, please provide a few sentences describing whether the model $y = w_0 + w_1 x$ is a good fit to the data, if $x$ represents earthquake magnitude and $y$ represents relative frequency.**

Answer: The model $y = w_0 + w_1 x$ is not a good fit to the data.  As the magitude (x) increases, the relative frequency (y) tends quickly towards 0.  This is characteristic of exponential decay, where there is a horizontal asymptote at $y = 0$.  This is not characteristic of a linear model in which $\frac{\Delta y}{\Delta x}$ is uniform for all pairs of points (x,y).

```python
def centroid(freq, min_val, max_val, data):
    tot = 0;
    for value in data:
        if(value >= min_val and value < max_val):
            tot += value
    
    if(freq != 0):
        return tot / freq
    else:
        return (min_val + max_val)/2

centroids = []
for i in range(len(h_bins)-1):
    centroids.append(centroid(hist[0][i], hist[1][i], hist[1][i+1], x))

plt.xlabel('Bin Centroid Magnitude')
plt.ylabel('Frequency')
plt.title('Frequency of Earthquakes by Binned Centroid')
plt.plot(centroids, hist[0], 'go')
plt.show()
```

As it turns out, a physical model that gives the predicted relative frequency of earthquakes of different magnitudes can be derived from physics, and is given by a so-called *power-law*

y = ax^b

**(20pts) If you wanted to directly fit this power-law model to these data, would you be able to use the linear regression code that we've already developed to do so?  If so, how?  If not, why not?**

Answer: No, you would not be able to directly fit this power-law model to these data with the linear regression code we've already developed.  In order to do this we must linearize the data (i.e. take the log transform of the data).  This will produce a linear relationship, and allow us to fit a line to the data.  To find the predicted value of the untransformed data we simply take $10^y$ where $y$ is the predicted value after fitting the log transformed data with linear regression. 

As it turns out, it is not possible to use linear regression to fit this model directly.  However, it can be done by using a clever transformation of the data.  **(20pts) Plot the base-10 logarithm of the event counts.  Now does a linear model seem like a good fit?  If the power law is a good fit to the untransformed data, why does it make sense that a linear model should fit the log-transformed data (HINT: take the logarithm of both sides of the power law)**

Answer: Yes, now a linear model seems like a good fit.  Taking the $\log_{10}$ of both sides we get the following..
$$\log_{10}{y}=\log_{10}{ax^b}$$

$$\log{10}{y}=b\log_{10}{ax}$$

$$\log{10}{y}=b\log_{10}{a}+\log{10}{x}$$
Which is linearized, (i.e. in a linear form $y = ax + b$).

```python
plt.title('Log10 Frequency and Log10 Bin Centroid Magnitude')
plt.xlabel('Log10 Bin Centroid Magnitude')
plt.ylabel('Log10 Frequency')
freq = hist[0]
x = []
y = []
for i in range(len(freq)):
    if(freq[i] != 0):
        x.append(centroids[i])
        y.append(freq[i])
x = np.array(x)
y = np.array(y)
plt.plot(x, np.log10(y), 'go')
plt.show()
```

Finally, **(40pts) use one of the methods that you developed for linear regression in class to fit a line to the log-transformed counts.  What is the slope of that line?  Is the model a good fit for all sizes of earthquake?**

Answer: The slope of the line is approximately $-0.957$ as shown below in the diagram.  The model is a good fit for the majority of sizes of earthquakes (as shown in the second figure below).  However, there are a few magnitudes where the model doesn't predict the observed values very well.  These are shown by gaps in the observed (blue), and predicted (red).

```python
def lbf(x_data, y):
    X = []
    for x in range(len(x_data)):
        X.append([1, x_data[x]])

    X = np.array(X)
    A = np.transpose(X)@X
    B = np.transpose(X)@y
    return np.linalg.solve(A,B)

est = lbf(x,np.log10(y))
plt.title("Figure 1. Log Transformed Data")
plt.xlabel("Log10 Bin Centroid Magnitude")
plt.ylabel("Log10 Freqency")
plt.plot(x, est[0] + est[1]*x, 'r-')
plt.plot(x, np.log10(y), 'go')
plt.text(3,1, "Slope: {}".format(round(est[1], 3)))
plt.text(3,.75, "Intercept: {}".format(round(est[0], 3)))
plt.show()
plt.title("Figure 2. How good is the model?(red)")
plt.xlabel("Magnitude")
plt.ylabel("Frequency")
plt.plot(x, y, 'b-')
plt.plot(x, 10**(est[0] + est[1]*x), 'r-')
plt.show()
```
